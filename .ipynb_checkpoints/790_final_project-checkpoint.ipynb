## Configure Packages

```python
import pandas as pd
import numpy as np
import sklearn # machine learning package

# data processing and cross validation
from sklearn import linear_model, preprocessing
from sklearn.preprocessing import StandardScaler, MinMaxScaler # scaling package
from sklearn.linear_model import SGDClassifier, Lasso # lasso regression and SGDClassifier
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold, StratifiedShuffleSplit, RandomizedSearchCV  # cross validation and grid search 

# models 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB # guassian navie bayes 
from sklearn.linear_model import LogisticRegression # logistic regression
from sklearn.tree import DecisionTreeClassifier # decision tree
from sklearn import tree as DecisionTree
from sklearn import datasets
from sklearn.ensemble import RandomForestClassifier # random forest
from sklearn import svm# %pip install xgboost
import xgboost as xgb

# metrics
from sklearn.inspection import permutation_importance # permutation importance 
from sklearn import metrics # useful metrics for the classification model
from sklearn.metrics import accuracy_score, roc_auc_score, f1_score, confusion_matrix, classification_report, auc, roc_curve, RocCurveDisplay # metrics

# plot
from matplotlib import pyplot as plt
import seaborn as sns # plotting package
```

## Read Data

```python
# read text file into pandas DataFrame
X_train_cy = pd.read_csv("titanic_train.txt", sep=" ",header = None)
# display DataFrame
X_train_cy.columns =['class', 'age', 'sex']
X_train_cy=X_train_cy.applymap(str)
X_train_cy.dtypes
```

```python
X_test_cy = pd.read_csv("titanic_test.txt", sep=" ",header = None)
X_test_cy.columns =['class', 'age', 'sex']
X_test_cy=X_test_cy.applymap(str)
print(X_test_cy)
X_test_cy.dtypes
```

```python
y_test_cy = pd.read_csv("titanic_test.Y.txt", sep=" ", header = None)
y_test_cy.columns = ['dead', 'survive']
y_test_cy = y_test_cy["survive"]
print(y_test_cy)
```

```python
y_train_cy = pd.read_csv("titanic_train.Y.txt", sep=" ", header = None)
y_train_cy.columns = ['dead', 'survive']
y_train_cy = y_train_cy["survive"]
print(y_train_cy)
```

```python
one_hot_data = pd.get_dummies(X_train_cy[['class','age','sex']],drop_first=True)
one_hot_data_test = pd.get_dummies(X_test_cy[['class','age','sex']],drop_first=True)
```

```python
one_hot_data_test
```

```python
# read results from BRL
pred_brl_test = pd.read_csv("preds_fullpost_test.csv", header = None)
pred_brl_train = pd.read_csv("preds_fullpost_train.csv", header = None)
```

## BRL

```python
pred_brl_test = np.array(list(pred_brl_test))
preds_brl = []
for i in pred_brl_test:
    if(i < 0.5):
        preds_brl.append(0)
    else:
        preds_brl.append(1)
        
metrics.f1_score(y_test_cy, preds_brl, labels = [0,1])
```

## Three Layer neural network

```python
from numpy import loadtxt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
from keras import backend as K

def recall_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    possible_positives = K.sum(K.round(K.clip(y_true, 0, 1)))
    recall = true_positives / (possible_positives + K.epsilon())
    return recall

def precision_m(y_true, y_pred):
    true_positives = K.sum(K.round(K.clip(y_true * y_pred, 0, 1)))
    predicted_positives = K.sum(K.round(K.clip(y_pred, 0, 1)))
    precision = true_positives / (predicted_positives + K.epsilon())
    return precision

def f1_m(y_true, y_pred):
    precision = precision_m(y_true, y_pred)
    recall = recall_m(y_true, y_pred)
    return 2*((precision*recall)/(precision+recall+K.epsilon()))
```

```python
%%time
#  fully-connected network structure with three layers Dense
# split into input (X) and output (y) variables
X = one_hot_data
y = y_train_cy
# define the keras model
model = Sequential()
model.add(Dense(12, input_shape=(5,), activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy',f1_m,precision_m, recall_m])
# fit the keras model on the dataset
model.fit(X, y, epochs=150, batch_size=10)
# evaluate the keras model
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
loss, accuracy, f1_score, precision, recall = model.evaluate(one_hot_data_test, y_test_cy,verbose=0)
print('Accuracy: %.2f' % (accuracy*100))
print('F1 score: %.2f' % (f1_score*100))
```
